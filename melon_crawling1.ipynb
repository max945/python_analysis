```python
from selenium import webdriver
driver = webdriver.Chrome(r'C:\Users\max94\chapapa_practice\chromedriver_win32\chromedriver.exe')
url = 'http://www.melon.com/chart/index.htm'
driver.get(url)
from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

songs = soup.select('#lst50')
print(len(songs))
print(songs[0])

song = songs[0]
title = song.select('a')
len(title)

title = song.select('span>a')
len(title)

title = song.select('div.ellipsis.rank01>span>a')
len(title)

title = song.select('div.ellipsis.rank01>span>a')[0].text
title

singer = song.select('div.ellipsis.rank02>a')
len(singer)

singer = song.select('div.ellipsis.rank02>a')[0].text
singer

for song in songs:
    title = song.select('div.ellipsis.rank01>span>a')[0].text
    singer = song.select('div.ellipsis.rank02>a')[0].text
    print(title, singer, sep ='|')
```
